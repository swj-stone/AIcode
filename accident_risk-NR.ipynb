```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
import pandas as pd
import os

# 1. 准备数据
print("准备数据...")

# 分离特征和目标变量
if 'accident_risk' in train_encoded.columns:
    feature_cols = [col for col in train_encoded.columns if col != 'accident_risk']
    
    X_train = train_encoded[feature_cols].values
    y_train = train_encoded['accident_risk'].values
    
    X_test = test_encoded[feature_cols].values
    
    # 检查测试集是否有id列
    if 'id' in test_encoded.columns:
        test_ids = test_encoded['id'].values
    else:
        # 如果没有id列，创建从1开始的id
        test_ids = np.arange(1, len(X_test) + 1)
    
    has_test_target = 'accident_risk' in test_encoded.columns
    if has_test_target:
        y_test = test_encoded['accident_risk'].values
else:
    raise ValueError("训练数据中未找到'accident_risk'列")

print(f"训练集形状: X_train={X_train.shape}, y_train={y_train.shape}")
print(f"测试集形状: X_test={X_test.shape}")

# 2. 数据标准化
print("\n数据标准化...")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 转换为PyTorch张量
X_train_tensor = torch.FloatTensor(X_train_scaled)
y_train_tensor = torch.FloatTensor(y_train).reshape(-1, 1)

if has_test_target:
    X_test_tensor = torch.FloatTensor(X_test_scaled)
    y_test_tensor = torch.FloatTensor(y_test).reshape(-1, 1)

# 3. 定义神经网络模型
class RegressionNN(nn.Module):
    def __init__(self, input_dim):
        super(RegressionNN, self).__init__()
        self.linear = nn.Linear(input_dim, 1)
    
    def forward(self, x):
        return self.linear(x)

# 4. 使用10个子集训练模型
print("\n使用10个子集训练模型...")

input_dim = X_train.shape[1]
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"使用设备: {device}")

# 训练参数
n_subsets = 10  # 使用10个子集
subset_size = min(5000, len(X_train_scaled))  # 每个子集大小
num_epochs = 50
batch_size = 64

# 存储所有训练好的模型
trained_models = []

for subset_idx in range(n_subsets):
    print(f"\n训练子集 {subset_idx + 1}/{n_subsets}...")
    
    # 随机选择子集
    if len(X_train_scaled) > subset_size:
        indices = np.random.choice(len(X_train_scaled), subset_size, replace=False)
        X_subset = X_train_scaled[indices]
        y_subset = y_train[indices]
    else:
        X_subset = X_train_scaled
        y_subset = y_train
    
    # 转换为PyTorch张量
    X_tensor = torch.FloatTensor(X_subset)
    y_tensor = torch.FloatTensor(y_subset).reshape(-1, 1)
    
    # 创建数据加载器
    train_dataset = TensorDataset(X_tensor, y_tensor)
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    
    # 初始化模型
    model = RegressionNN(input_dim).to(device)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    
    # 训练模型
    model.train()
    for epoch in range(num_epochs):
        epoch_loss = 0
        for batch_X, batch_y in train_loader:
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y)
            
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            epoch_loss += loss.item()
        
        if (epoch + 1) % 10 == 0:
            print(f'  Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss/len(train_loader):.4f}')
    
    # 保存训练好的模型
    trained_models.append(model)
    
    # 在训练子集上评估
    model.eval()
    with torch.no_grad():
        X_subset_tensor = torch.FloatTensor(X_subset).to(device)
        subset_pred = model(X_subset_tensor).cpu().numpy()
        subset_r2 = r2_score(y_subset.flatten(), subset_pred.flatten())
        print(f"  子集评估 - R²: {subset_r2:.4f}")

# 5. 用所有训练好的模型对整个train集进行预测
print("\n使用训练好的模型对整个训练集进行预测...")

# 存储所有模型在完整训练集上的预测
all_train_preds = []

for i, model in enumerate(trained_models):
    model.eval()
    with torch.no_grad():
        X_train_tensor_device = X_train_tensor.to(device)
        train_pred = model(X_train_tensor_device).cpu().numpy()
        all_train_preds.append(train_pred.flatten())
    
    # 计算单个模型在完整训练集上的R²
    model_r2 = r2_score(y_train.flatten(), train_pred.flatten())
    print(f"  模型 {i+1} 在完整训练集上的R²: {model_r2:.4f}")

# 集成预测（取平均值）
print("\n集成所有模型的预测结果...")
all_train_preds = np.array(all_train_preds)
ensemble_train_pred = np.mean(all_train_preds, axis=0)

# 6. 计算集成模型在完整训练集上的准确率
print("\n计算集成模型在完整训练集上的准确率...")

# 计算集成模型在完整训练集上的指标
train_r2 = r2_score(y_train.flatten(), ensemble_train_pred.flatten())
train_mse = np.mean((y_train.flatten() - ensemble_train_pred.flatten()) ** 2)
train_mae = np.mean(np.abs(y_train.flatten() - ensemble_train_pred.flatten()))

print(f"集成模型在完整训练集上的评估:")
print(f"  R²分数: {train_r2:.4f}")
print(f"  均方误差 (MSE): {train_mse:.4f}")
print(f"  平均绝对误差 (MAE): {train_mae:.4f}")

# 7. 用训练好的模型对test集进行预测
print("\n使用训练好的模型对测试集进行预测...")

# 存储所有模型在测试集上的预测
all_test_preds = []

for i, model in enumerate(trained_models):
    model.eval()
    with torch.no_grad():
        X_test_tensor_device = torch.FloatTensor(X_test_scaled).to(device)
        test_pred = model(X_test_tensor_device).cpu().numpy()
        all_test_preds.append(test_pred.flatten())

# 集成预测（取平均值）
all_test_preds = np.array(all_test_preds)
ensemble_test_pred = np.mean(all_test_preds, axis=0)

# 8. 如果测试集有目标变量，计算最终评估指标
if has_test_target:
    print("\n集成模型在测试集上的评估:")
    print("-" * 50)
    
    test_r2 = r2_score(y_test.flatten(), ensemble_test_pred.flatten())
    test_mse = np.mean((y_test.flatten() - ensemble_test_pred.flatten()) ** 2)
    test_mae = np.mean(np.abs(y_test.flatten() - ensemble_test_pred.flatten()))
    
    print(f"  R²分数: {test_r2:.4f}")
    print(f"  均方误差 (MSE): {test_mse:.4f}")
    print(f"  平均绝对误差 (MAE): {test_mae:.4f}")
    
    # 显示预测与真实值的对比
    print("\n预测与真实值对比 (前10个样本):")
    print("-" * 50)
    for i in range(min(10, len(y_test))):
        print(f"  样本 {i+1}: 真实值={y_test[i]:.4f}, 预测值={ensemble_test_pred[i]:.4f}, 误差={abs(y_test[i]-ensemble_test_pred[i]):.4f}")
else:
    print("\n测试集没有目标变量，无法计算评估指标")

# 9. 创建输出文件
print("\n创建输出文件...")

# 创建结果DataFrame
result_df = pd.DataFrame({
    'id': test_ids,
    'pre': ensemble_test_pred
})

# 保存为CSV文件
output_path = os.path.join(base_path, "predictions.csv")
result_df.to_csv(output_path, index=False)
print(f"预测结果已保存到: {output_path}")
print(f"文件包含 {len(result_df)} 行数据")

# 10. 显示输出文件的前几行
print("\n输出文件前10行:")
print(result_df.head(10))

# 11. 显示预测统计信息
print("\n测试集预测统计信息:")
print("-" * 50)
print(f"预测值范围: [{ensemble_test_pred.min():.4f}, {ensemble_test_pred.max():.4f}]")
print(f"预测平均值: {ensemble_test_pred.mean():.4f}")
print(f"预测标准差: {ensemble_test_pred.std():.4f}")

# 12. 总结报告
print("\n" + "="*60)
print("训练总结报告")
print("="*60)
print(f"训练策略: 使用{n_subsets}个子集，每个子集{subset_size}个样本")
print(f"训练轮数: {num_epochs} epochs")
print(f"批量大小: {batch_size}")
print(f"训练设备: {device}")
print(f"训练集最终R²: {train_r2:.4f}")
if has_test_target:
    print(f"测试集最终R²: {test_r2:.4f}")
print(f"预测文件: {output_path}")
print("="*60)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
import pandas as pd
import os

# 1. 准备数据
print("准备数据...")

# 分离特征和目标变量
if 'accident_risk' in train_encoded.columns:
    feature_cols = [col for col in train_encoded.columns if col != 'accident_risk']
    
    # 保存目标变量用于最后计算准确率
    y_train = train_encoded['accident_risk'].values.copy()  # 保存目标变量
    y_train_original = y_train.copy()  # 保存原始值用于评估
    
    # 从训练数据中移除accident_risk列
    X_train = train_encoded[feature_cols].values
    
    X_test = test_encoded[feature_cols].values
    
    # 检查测试集是否有id列
    if 'id' in test_encoded.columns:
        test_ids = test_encoded['id'].values
    else:
        # 如果没有id列，创建从1开始的id
        test_ids = np.arange(1, len(X_test) + 1)
    
    has_test_target = 'accident_risk' in test_encoded.columns
    if has_test_target:
        y_test = test_encoded['accident_risk'].values
        y_test_original = y_test.copy()  # 保存测试集目标变量
else:
    raise ValueError("训练数据中未找到'accident_risk'列")

print(f"训练集形状: X_train={X_train.shape}, y_train={len(y_train)}")
print(f"测试集形状: X_test={X_test.shape}")
print(f"已从训练特征中移除'accident_risk'列")

# 2. 数据标准化
print("\n数据标准化...")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 转换为PyTorch张量
X_train_tensor = torch.FloatTensor(X_train_scaled)
y_train_tensor = torch.FloatTensor(y_train).reshape(-1, 1)

if has_test_target:
    X_test_tensor = torch.FloatTensor(X_test_scaled)
    y_test_tensor = torch.FloatTensor(y_test).reshape(-1, 1)

# 3. 定义神经网络模型
class RegressionNN(nn.Module):
    def __init__(self, input_dim):
        super(RegressionNN, self).__init__()
        self.linear = nn.Linear(input_dim, 1)
    
    def forward(self, x):
        return self.linear(x)

# 4. 使用10个子集训练模型
print("\n使用10个子集训练模型...")

input_dim = X_train.shape[1]
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"使用设备: {device}")

# 训练参数
n_subsets = 10  # 使用10个子集
subset_size = min(5000, len(X_train_scaled))  # 每个子集大小
num_epochs = 50
batch_size = 64

# 存储所有训练好的模型
trained_models = []

for subset_idx in range(n_subsets):
    print(f"\n训练子集 {subset_idx + 1}/{n_subsets}...")
    
    # 随机选择子集
    if len(X_train_scaled) > subset_size:
        indices = np.random.choice(len(X_train_scaled), subset_size, replace=False)
        X_subset = X_train_scaled[indices]
        y_subset = y_train[indices]
    else:
        X_subset = X_train_scaled
        y_subset = y_train
    
    # 转换为PyTorch张量
    X_tensor = torch.FloatTensor(X_subset)
    y_tensor = torch.FloatTensor(y_subset).reshape(-1, 1)
    
    # 创建数据加载器
    train_dataset = TensorDataset(X_tensor, y_tensor)
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    
    # 初始化模型
    model = RegressionNN(input_dim).to(device)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    
    # 训练模型
    model.train()
    for epoch in range(num_epochs):
        epoch_loss = 0
        for batch_X, batch_y in train_loader:
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y)
            
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            epoch_loss += loss.item()
        
        if (epoch + 1) % 10 == 0:
            print(f'  Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss/len(train_loader):.4f}')
    
    # 保存训练好的模型
    trained_models.append(model)
    
    # 在训练子集上评估
    model.eval()
    with torch.no_grad():
        X_subset_tensor = torch.FloatTensor(X_subset).to(device)
        subset_pred = model(X_subset_tensor).cpu().numpy()
        subset_r2 = r2_score(y_subset.flatten(), subset_pred.flatten())
        print(f"  子集评估 - R²: {subset_r2:.4f}")

# 5. 用所有训练好的模型对整个train集进行预测
print("\n使用训练好的模型对整个训练集进行预测...")

# 存储所有模型在完整训练集上的预测
all_train_preds = []

for i, model in enumerate(trained_models):
    model.eval()
    with torch.no_grad():
        X_train_tensor_device = X_train_tensor.to(device)
        train_pred = model(X_train_tensor_device).cpu().numpy()
        all_train_preds.append(train_pred.flatten())
    
    # 计算单个模型在完整训练集上的R²
    model_r2 = r2_score(y_train.flatten(), train_pred.flatten())
    print(f"  模型 {i+1} 在完整训练集上的R²: {model_r2:.4f}")

# 集成预测（取平均值）
print("\n集成所有模型的预测结果...")
all_train_preds = np.array(all_train_preds)
ensemble_train_pred = np.mean(all_train_preds, axis=0)

# 6. 计算集成模型在完整训练集上的准确率（使用保存的y_train）
print("\n计算集成模型在完整训练集上的准确率...")

# 计算集成模型在完整训练集上的指标
train_r2 = r2_score(y_train_original.flatten(), ensemble_train_pred.flatten())
train_mse = np.mean((y_train_original.flatten() - ensemble_train_pred.flatten()) ** 2)
train_mae = np.mean(np.abs(y_train_original.flatten() - ensemble_train_pred.flatten()))

print(f"集成模型在完整训练集上的评估:")
print(f"  R²分数: {train_r2:.4f}")
print(f"  均方误差 (MSE): {train_mse:.4f}")
print(f"  平均绝对误差 (MAE): {train_mae:.4f}")

# 7. 用训练好的模型对test集进行预测
print("\n使用训练好的模型对测试集进行预测...")

# 存储所有模型在测试集上的预测
all_test_preds = []

for i, model in enumerate(trained_models):
    model.eval()
    with torch.no_grad():
        X_test_tensor_device = torch.FloatTensor(X_test_scaled).to(device)
        test_pred = model(X_test_tensor_device).cpu().numpy()
        all_test_preds.append(test_pred.flatten())

# 集成预测（取平均值）
all_test_preds = np.array(all_test_preds)
ensemble_test_pred = np.mean(all_test_preds, axis=0)

# 8. 如果测试集有目标变量，计算最终评估指标
if has_test_target:
    print("\n集成模型在测试集上的评估:")
    print("-" * 50)
    
    test_r2 = r2_score(y_test_original.flatten(), ensemble_test_pred.flatten())
    test_mse = np.mean((y_test_original.flatten() - ensemble_test_pred.flatten()) ** 2)
    test_mae = np.mean(np.abs(y_test_original.flatten() - ensemble_test_pred.flatten()))
    
    print(f"  R²分数: {test_r2:.4f}")
    print(f"  均方误差 (MSE): {test_mse:.4f}")
    print(f"  平均绝对误差 (MAE): {test_mae:.4f}")
    
    # 显示预测与真实值的对比
    print("\n预测与真实值对比 (前10个样本):")
    print("-" * 50)
    for i in range(min(10, len(y_test_original))):
        print(f"  样本 {i+1}: 真实值={y_test_original[i]:.4f}, 预测值={ensemble_test_pred[i]:.4f}, 误差={abs(y_test_original[i]-ensemble_test_pred[i]):.4f}")
else:
    print("\n测试集没有目标变量，无法计算评估指标")

# 9. 创建输出文件
print("\n创建输出文件...")

# 创建结果DataFrame
result_df = pd.DataFrame({
    'id': test_ids,
    'pre': ensemble_test_pred
})

# 保存为CSV文件
output_path = os.path.join(base_path, "predictions.csv")
result_df.to_csv(output_path, index=False)
print(f"预测结果已保存到: {output_path}")
print(f"文件包含 {len(result_df)} 行数据")

# 10. 显示输出文件的前几行
print("\n输出文件前10行:")
print(result_df.head(10))

# 11. 显示预测统计信息
print("\n测试集预测统计信息:")
print("-" * 50)
print(f"预测值范围: [{ensemble_test_pred.min():.4f}, {ensemble_test_pred.max():.4f}]")
print(f"预测平均值: {ensemble_test_pred.mean():.4f}")
print(f"预测标准差: {ensemble_test_pred.std():.4f}")

# 12. 总结报告
print("\n" + "="*60)
print("训练总结报告")
print("="*60)
print(f"训练策略: 使用{n_subsets}个子集，每个子集{subset_size}个样本")
print(f"训练轮数: {num_epochs} epochs")
print(f"批量大小: {batch_size}")
print(f"训练设备: {device}")
print(f"训练集最终R²: {train_r2:.4f}")
if has_test_target:
    print(f"测试集最终R²: {test_r2:.4f}")
print(f"预测文件: {output_path}")
print("="*60)

# 13. 验证数据完整性
print("\n数据完整性验证:")
print("-" * 50)
print(f"训练特征维度: {X_train.shape[1]} 个特征")
print(f"训练目标变量长度: {len(y_train_original)}")
print(f"训练特征与目标变量是否匹配: {X_train.shape[0] == len(y_train_original)}")
print(f"accident_risk是否已从训练特征中移除: {'accident_risk' not in feature_cols}")
print(f"预测结果数量: {len(ensemble_test_pred)}")
print(f"测试ID数量: {len(test_ids)}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
import pandas as pd
import os

# 第一步：基于已独热编码的数据集进行特征杂交
print("第一步：基于已独热编码的数据集进行特征杂交")
print("="*60)

# 1.1 复制已处理好的数据集
print("复制已独热编码的数据集...")
train_encoded_copy = train_encoded.copy()
test_encoded_copy = test_encoded.copy()

print(f"训练集副本形状: {train_encoded_copy.shape}")
print(f"测试集副本形状: {test_encoded_copy.shape}")

# 1.2 准备特征数据（排除目标列和ID列）
print("\n准备特征数据...")

# 保存目标变量（仅训练集有）
if 'accident_risk' in train_encoded_copy.columns:
    y_train_original = train_encoded_copy['accident_risk'].values.copy()
    
    # 从训练数据中移除accident_risk列
    feature_cols = [col for col in train_encoded_copy.columns if col != 'accident_risk']
    
    # 检查是否有ID列，如果有也排除
    if 'id' in feature_cols:
        feature_cols.remove('id')
        train_ids = train_encoded_copy['id'].values if 'id' in train_encoded_copy.columns else None
    else:
        train_ids = None
    
    # 获取训练集特征数据
    X_train_original = train_encoded_copy[feature_cols].values
    
    print(f"原始特征数量: {len(feature_cols)}")
    print(f"原始特征矩阵形状 - 训练集: {X_train_original.shape}")
    print(f"训练集目标变量形状: {y_train_original.shape}")
else:
    raise ValueError("训练数据中未找到'accident_risk'列")

# 处理测试集（没有accident_risk目标变量）
# 检查测试集是否有ID列
if 'id' in test_encoded_copy.columns:
    test_ids = test_encoded_copy['id'].values
    # 从测试数据中移除id列（如果存在）
    test_feature_cols = [col for col in test_encoded_copy.columns if col != 'id']
else:
    test_ids = np.arange(1, len(test_encoded_copy)+1)
    test_feature_cols = test_encoded_copy.columns.tolist()

# 确保测试集特征列与训练集一致
print(f"\n确保测试集特征与训练集一致...")
missing_in_test = set(feature_cols) - set(test_feature_cols)
extra_in_test = set(test_feature_cols) - set(feature_cols)

if missing_in_test:
    print(f"  测试集缺少列: {len(missing_in_test)}个")
    for col in missing_in_test:
        test_encoded_copy[col] = 0

if extra_in_test:
    print(f"  测试集多余列: {len(extra_in_test)}个")
    # 移除测试集多余列
    test_encoded_copy = test_encoded_copy.drop(columns=list(extra_in_test))

# 获取测试集特征数据
X_test_original = test_encoded_copy[feature_cols].values
print(f"原始特征矩阵形状 - 测试集: {X_test_original.shape}")

# 测试集没有目标变量
has_test_target = False
y_test_original = None
print("测试集没有目标变量，仅用于预测")

# 1.3 特征杂交函数（生成所有特征的二次项：平方和乘积）
def create_quadratic_features_matrix(X, feature_names=None):
    """
    创建二次杂交特征矩阵
    包含：1. 原始特征 2. 每个特征的平方 3. 每对特征的乘积
    """
    n_samples, n_features = X.shape
    print(f"原始特征数: {n_features}")
    
    # 计算二次特征数量
    # 自身平方: n_features个
    # 两两组合: C(n, 2) = n*(n-1)/2 个
    n_square = n_features
    n_product = n_features * (n_features - 1) // 2
    n_total_features = n_features + n_square + n_product
    
    print(f"预计创建特征:")
    print(f"  - 原始特征: {n_features}")
    print(f"  - 平方特征: {n_square}")
    print(f"  - 乘积特征: {n_product}")
    print(f"  - 总计: {n_total_features}")
    
    # 创建新的特征矩阵
    X_quadratic = np.zeros((n_samples, n_total_features))
    
    # 1. 复制原始特征
    X_quadratic[:, :n_features] = X.copy()
    print("  已复制原始特征")
    
    # 2. 创建平方特征
    current_idx = n_features
    for i in range(n_features):
        X_quadratic[:, current_idx] = X[:, i] * X[:, i]
        current_idx += 1
    print(f"  已创建{current_idx - n_features}个平方特征")
    
    # 3. 创建乘积特征
    progress_interval = max(1, n_product // 10)  # 每10%显示一次进度
    product_count = 0
    
    for i in range(n_features):
        for j in range(i+1, n_features):
            X_quadratic[:, current_idx] = X[:, i] * X[:, j]
            current_idx += 1
            product_count += 1
            
            if product_count % progress_interval == 0:
                progress = product_count / n_product * 100
                print(f"  已创建{product_count}/{n_product}个乘积特征 ({progress:.1f}%)")
    
    print(f"特征杂交完成！最终特征数: {X_quadratic.shape[1]}")
    
    return X_quadratic

# 1.4 为训练集和测试集创建二次杂交特征
print("\n为训练集创建二次杂交特征...")
X_train_quadratic = create_quadratic_features_matrix(X_train_original)

print("\n为测试集创建二次杂交特征...")
X_test_quadratic = create_quadratic_features_matrix(X_test_original)

print(f"\n二次杂交后特征矩阵形状:")
print(f"  训练集: {X_train_quadratic.shape}")
print(f"  测试集: {X_test_quadratic.shape}")

# 1.5 标准化二次杂交特征
print("\n标准化二次杂交特征...")
scaler_quadratic = StandardScaler()
X_train_scaled = scaler_quadratic.fit_transform(X_train_quadratic)
X_test_scaled = scaler_quadratic.transform(X_test_quadratic)

print(f"标准化后特征矩阵形状:")
print(f"  训练集: {X_train_scaled.shape}")
print(f"  测试集: {X_test_scaled.shape}")

# 第二步：拆分训练集为训练和验证集
print("\n" + "="*60)
print("第二步：拆分训练集为训练和验证集")
print("="*60)

# 设置随机种子以确保可重复性
np.random.seed(42)
torch.manual_seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed(42)

# 2.1 拆分训练集为训练集和验证集（80%训练，20%验证）
n_samples = len(X_train_scaled)
indices = np.arange(n_samples)
np.random.shuffle(indices)

train_ratio = 0.8
train_size = int(n_samples * train_ratio)

train_indices = indices[:train_size]
val_indices = indices[train_size:]

X_train_split = X_train_scaled[train_indices]
y_train_split = y_train_original[train_indices]
X_val_split = X_train_scaled[val_indices]
y_val_split = y_train_original[val_indices]

print(f"训练集拆分结果:")
print(f"  总样本数: {n_samples}")
print(f"  训练集样本数: {len(X_train_split)} ({train_ratio*100:.0f}%)")
print(f"  验证集样本数: {len(X_val_split)} ({100 - train_ratio*100:.0f}%)")
print(f"  训练集特征形状: {X_train_split.shape}")
print(f"  验证集特征形状: {X_val_split.shape}")

# 转换为PyTorch张量
X_train_tensor = torch.FloatTensor(X_train_split)
y_train_tensor = torch.FloatTensor(y_train_split).reshape(-1, 1)
X_val_tensor = torch.FloatTensor(X_val_split)
y_val_tensor = torch.FloatTensor(y_val_split).reshape(-1, 1)

# 第三步：使用训练集训练模型，用验证集验证
print("\n" + "="*60)
print("第三步：使用训练集训练模型，用验证集验证")
print("="*60)

# 3.1 定义神经网络模型（针对高维特征优化）
class QuadraticFeatureNN(nn.Module):
    def __init__(self, input_dim):
        super(QuadraticFeatureNN, self).__init__()
        # 针对高维特征的网络结构
        self.dropout_input = nn.Dropout(0.4)  # 输入层高dropout
        self.linear1 = nn.Linear(input_dim, 128)
        self.bn1 = nn.BatchNorm1d(128)  # 批归一化
        self.dropout1 = nn.Dropout(0.3)
        self.linear2 = nn.Linear(128, 64)
        self.bn2 = nn.BatchNorm1d(64)
        self.dropout2 = nn.Dropout(0.3)
        self.linear3 = nn.Linear(64, 32)
        self.linear4 = nn.Linear(32, 1)
        self.relu = nn.ReLU()
    
    def forward(self, x):
        x = self.dropout_input(x)
        x = self.relu(self.bn1(self.linear1(x)))
        x = self.dropout1(x)
        x = self.relu(self.bn2(self.linear2(x)))
        x = self.dropout2(x)
        x = self.relu(self.linear3(x))
        x = self.linear4(x)
        return x

# 3.2 使用10个子集训练模型（使用训练集部分）
print("\n使用10个子集训练二次杂交特征模型...")

input_dim = X_train_scaled.shape[1]
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"使用设备: {device}")
print(f"输入特征维度: {input_dim}")

# 训练参数调整（针对高维特征）
n_subsets = 10
subset_size = min(4000, len(X_train_split))  # 使用训练集部分
num_epochs = 60  # 增加训练轮数
batch_size = 128  # 增加批量大小

# 存储所有训练好的模型
trained_models = []
val_performance = []  # 存储每个模型在验证集上的性能

for subset_idx in range(n_subsets):
    print(f"\n训练子集 {subset_idx + 1}/{n_subsets}...")
    
    # 随机选择子集（从训练集部分）
    if len(X_train_split) > subset_size:
        indices = np.random.choice(len(X_train_split), subset_size, replace=False)
        X_subset = X_train_split[indices]
        y_subset = y_train_split[indices]
    else:
        X_subset = X_train_split
        y_subset = y_train_split
    
    # 转换为PyTorch张量
    X_tensor = torch.FloatTensor(X_subset)
    y_tensor = torch.FloatTensor(y_subset).reshape(-1, 1)
    
    # 创建数据加载器
    train_dataset = TensorDataset(X_tensor, y_tensor)
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    
    # 初始化模型
    model = QuadraticFeatureNN(input_dim).to(device)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.0005, weight_decay=0.02)  # 降低学习率，增加L2正则化
    
    # 学习率调度器
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5)
    
    # 训练模型
    model.train()
    best_loss = float('inf')
    patience_counter = 0
    
    for epoch in range(num_epochs):
        epoch_loss = 0
        for batch_X, batch_y in train_loader:
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y)
            
            optimizer.zero_grad()
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)  # 梯度裁剪
            optimizer.step()
            
            epoch_loss += loss.item()
        
        avg_loss = epoch_loss / len(train_loader)
        scheduler.step(avg_loss)
        
        if (epoch + 1) % 10 == 0:
            print(f'  Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}, LR: {optimizer.param_groups[0]["lr"]:.6f}')
    
    # 保存训练好的模型
    trained_models.append(model)
    
    # 在训练子集上评估
    model.eval()
    with torch.no_grad():
        X_subset_tensor = torch.FloatTensor(X_subset).to(device)
        subset_pred = model(X_subset_tensor).cpu().numpy()
        subset_r2 = r2_score(y_subset.flatten(), subset_pred.flatten())
        print(f"  训练子集评估 - R²: {subset_r2:.4f}")
        
        # 在验证集上评估
        X_val_tensor_device = X_val_tensor.to(device)
        val_pred = model(X_val_tensor_device).cpu().numpy()
        val_r2 = r2_score(y_val_split.flatten(), val_pred.flatten())
        val_performance.append(val_r2)
        print(f"  验证集评估 - R²: {val_r2:.4f}")

# 3.3 用所有训练好的模型对整个训练集进行预测（包括训练集和验证集）
print("\n使用所有模型对整个训练集（训练+验证）进行预测...")

all_full_train_preds = []  # 对整个训练集（训练+验证）的预测

for i, model in enumerate(trained_models):
    model.eval()
    with torch.no_grad():
        X_full_train_tensor = torch.FloatTensor(X_train_scaled).to(device)
        full_train_pred = model(X_full_train_tensor).cpu().numpy()
        all_full_train_preds.append(full_train_pred.flatten())

# 集成预测（对整个训练集）
all_full_train_preds = np.array(all_full_train_preds)
ensemble_full_train_pred = np.mean(all_full_train_preds, axis=0)

# 3.4 计算集成模型在整个训练集上的准确率
print("\n计算集成模型在整个训练集（训练+验证）上的准确率...")

full_train_r2 = r2_score(y_train_original.flatten(), ensemble_full_train_pred.flatten())
full_train_mse = np.mean((y_train_original.flatten() - ensemble_full_train_pred.flatten()) ** 2)
full_train_mae = np.mean(np.abs(y_train_original.flatten() - ensemble_full_train_pred.flatten()))

print(f"集成模型在整个训练集上的评估:")
print(f"  R²分数: {full_train_r2:.4f}")
print(f"  均方误差 (MSE): {full_train_mse:.4f}")
print(f"  平均绝对误差 (MAE): {full_train_mae:.4f}")

# 3.5 计算集成模型在验证集上的准确率
print("\n计算集成模型在验证集上的准确率...")

# 用所有模型对验证集进行预测
all_val_preds = []

for i, model in enumerate(trained_models):
    model.eval()
    with torch.no_grad():
        X_val_tensor_device = X_val_tensor.to(device)
        val_pred = model(X_val_tensor_device).cpu().numpy()
        all_val_preds.append(val_pred.flatten())

# 集成预测（验证集）
all_val_preds = np.array(all_val_preds)
ensemble_val_pred = np.mean(all_val_preds, axis=0)

val_r2 = r2_score(y_val_split.flatten(), ensemble_val_pred.flatten())
val_mse = np.mean((y_val_split.flatten() - ensemble_val_pred.flatten()) ** 2)
val_mae = np.mean(np.abs(y_val_split.flatten() - ensemble_val_pred.flatten()))

print(f"集成模型在验证集上的评估:")
print(f"  R²分数: {val_r2:.4f}")
print(f"  均方误差 (MSE): {val_mse:.4f}")
print(f"  平均绝对误差 (MAE): {val_mae:.4f}")

# 3.6 用训练好的模型对test集进行预测
print("\n使用训练好的模型对测试集进行预测...")

all_test_preds = []

for i, model in enumerate(trained_models):
    model.eval()
    with torch.no_grad():
        X_test_tensor_device = torch.FloatTensor(X_test_scaled).to(device)
        test_pred = model(X_test_tensor_device).cpu().numpy()
        all_test_preds.append(test_pred.flatten())

# 集成预测（测试集）
all_test_preds = np.array(all_test_preds)
ensemble_test_pred = np.mean(all_test_preds, axis=0)

# 第四步：创建输出文件
print("\n" + "="*60)
print("第四步：创建输出文件")
print("="*60)

# 4.1 创建验证集预测文件
val_result_df = pd.DataFrame({
    'id': train_ids[val_indices] if train_ids is not None else val_indices,
    'true_value': y_val_split,
    'predicted_value': ensemble_val_pred
})

val_output_path = os.path.join(base_path, "validation_predictions_quadratic.csv")
val_result_df.to_csv(val_output_path, index=False)
print(f"验证集预测结果已保存到: {val_output_path}")
print(f"验证集文件包含 {len(val_result_df)} 行数据")

# 4.2 创建测试集预测文件
test_result_df = pd.DataFrame({
    'id': test_ids,
    'pre': ensemble_test_pred
})

test_output_path = os.path.join(base_path, "test_predictions_quadratic.csv")
test_result_df.to_csv(test_output_path, index=False)
print(f"测试集预测结果已保存到: {test_output_path}")
print(f"测试集文件包含 {len(test_result_df)} 行数据")

# 显示输出文件的前几行
print("\n验证集输出文件前5行:")
print(val_result_df.head(5))
print("\n测试集输出文件前5行:")
print(test_result_df.head(5))

# 第五步：性能分析与总结
print("\n" + "="*60)
print("第五步：性能分析与总结")
print("="*60)

# 5.1 特征分析
print(f"特征分析:")
print(f"  原始特征数: {len(feature_cols)}")
print(f"  二次杂交后特征数: {input_dim}")
print(f"  特征扩展倍数: {input_dim/len(feature_cols):.1f}x")

# 5.2 模型复杂度分析
print(f"\n模型复杂度分析:")
print(f"  网络结构: 输入层({input_dim}) -> 128 -> 64 -> 32 -> 1")
print(f"  正则化配置:")
print(f"    - Dropout: 输入层(0.4), 隐藏层(0.3)")
print(f"    - L2正则化: 0.02")
print(f"    - 批归一化: 是")
print(f"    - 梯度裁剪: 是")

# 5.3 训练和验证性能总结
print(f"\n性能总结:")
print(f"  训练集（80%）上训练的模型在验证集（20%）上的表现:")
print(f"    R²分数: {val_r2:.4f}")
print(f"    均方误差 (MSE): {val_mse:.4f}")
print(f"    平均绝对误差 (MAE): {val_mae:.4f}")

print(f"\n  集成模型在整个训练集（100%）上的表现:")
print(f"    R²分数: {full_train_r2:.4f}")
print(f"    均方误差 (MSE): {full_train_mse:.4f}")
print(f"    平均绝对误差 (MAE): {full_train_mae:.4f}")

# 5.4 过拟合分析
print(f"\n过拟合分析:")
overfit_gap = full_train_r2 - val_r2
print(f"  训练集与验证集R²差距: {overfit_gap:.4f}")

if overfit_gap < 0.05:
    print(f"  ✅ 优秀：几乎没有过拟合（差距 < 0.05）")
elif overfit_gap < 0.1:
    print(f"  ✓ 良好：轻微过拟合（差距 < 0.1）")
elif overfit_gap < 0.15:
    print(f"  ℹ 正常：中等程度过拟合（差距 < 0.15）")
elif overfit_gap < 0.2:
    print(f"  ⚠ 注意：较严重过拟合（差距 < 0.2）")
else:
    print(f"  ❗ 警告：严重过拟合（差距 ≥ 0.2）")

# 5.5 各模型验证集性能统计
print(f"\n各模型在验证集上的性能统计:")
print(f"  最佳模型R²: {max(val_performance):.4f} (模型{val_performance.index(max(val_performance))+1})")
print(f"  最差模型R²: {min(val_performance):.4f} (模型{val_performance.index(min(val_performance))+1})")
print(f"  平均R²: {np.mean(val_performance):.4f}")
print(f"  R²标准差: {np.std(val_performance):.4f}")

# 5.6 预测统计信息
print(f"\n测试集预测统计信息:")
print(f"  预测值范围: [{ensemble_test_pred.min():.4f}, {ensemble_test_pred.max():.4f}]")
print(f"  预测平均值: {ensemble_test_pred.mean():.4f}")
print(f"  预测标准差: {ensemble_test_pred.std():.4f}")

# 5.7 与原始模型对比
try:
    original_val_r2 = 0.7985  # 这里需要替换为原始模型在验证集上的实际R²
    original_full_train_r2 = 0.8072  # 原始模型在整个训练集上的R²
    
    print(f"\n与原始特征模型对比:")
    print(f"  原始模型 - 验证集R²: {original_val_r2:.4f}")
    print(f"  二次杂交模型 - 验证集R²: {val_r2:.4f}")
    
    if val_r2 > original_val_r2:
        improvement = val_r2 - original_val_r2
        improvement_percent = (improvement / abs(original_val_r2)) * 100 if original_val_r2 != 0 else 0
        print(f"  ✅ 二次杂交特征提升了模型验证性能")
        print(f"  验证集R²提升: +{improvement:.4f} ({improvement_percent:.1f}%)")
    else:
        print(f"  ⚠ 二次杂交特征没有提升验证集性能")
    
    print(f"\n  原始模型 - 完整训练集R²: {original_full_train_r2:.4f}")
    print(f"  二次杂交模型 - 完整训练集R²: {full_train_r2:.4f}")
    
except Exception as e:
    print(f"\n无法与原始模型进行完整对比: {e}")
    print("需要原始模型的验证集和完整训练集R²值进行对比")

# 5.8 最终总结
print(f"\n最终总结:")
if val_r2 > 0.8:
    print(f"  ✅ 优秀：验证集表现优异（R² > 0.8）")
elif val_r2 > 0.7:
    print(f"  ✓ 良好：验证集表现良好（R² > 0.7）")
elif val_r2 > 0.6:
    print(f"  ℹ 一般：验证集表现一般（R² > 0.6）")
elif val_r2 > 0.5:
    print(f"  ⚠ 需要改进：验证集表现不佳（R² > 0.5）")
else:
    print(f"  ❗ 差：验证集表现很差（R² ≤ 0.5）")

print(f"\n输出文件:")
print(f"  验证集预测文件: {val_output_path}")
print(f"  测试集预测文件: {test_output_path}")
print("="*60)
```
